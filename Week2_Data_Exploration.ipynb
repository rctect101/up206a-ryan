```python
#use the Latitude/Longitude column to make a map of housing starts

#import folium
import folium

#my house in mid city is about in the middle of LA
my_house = [34.047473, -118.332487]

#initialize map
m = folium.Map(location=my_house, zoom_start=10)

#add housing starts
for index, row in permits.iterrows():
    folium.CircleMarker((
        row['Latitude'], row['Longitude']),
        radius=1,
        fill_color="#3db7e4"
        ).add_to(m)

# convert to (n, 2) nd-array format for heatmap (from https://alysivji.github.io/getting-started-with-folium.html)
stationArr = permits[['Latitude', 'Longitude']].to_numpy()

# plot heatmap
from folium import plugins
from folium.plugins import HeatMap
m.add_child(plugins.HeatMap(stationArr, radius=20))

m
```
