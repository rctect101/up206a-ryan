```python
# import folium, a leaflet python library
import folium

#create a list of airport coordinates
akl = [(-37.007729, 174.789544), "Auckland", "AKL"]
atl = [(33.635012, -84.428656), "Atlanta", "ATL"]
dfw = [(32.897654, -97.050628), "Dallas", "DFW"]
doh = [(25.274909, 51.614785), "Doha", "DOH"]
dxb = [(25.252390, 55.364430), "Dubai", "DXB"]
ewr = [(40.688725, -74.173570),"Newark","EWR"]
iah = [(29.990996, -95.343360), "Houston", "IAH"]
jfk = [(40.641189, -73.776830), "New York", "JFK"]
jnb = [(-26.133957, 28.237802), "Johannesburg", "JNB"]
lax = [(33.942528, -118.414792), "Los Angeles", "LAX"]
lhr = [(51.471734, -0.467853), "London", "LHR"]
mnl = [(14.510259, 121.018171), "Manila", "MNL"]
per = [(-31.943535, 115.964169), "Perth", "PER"]
sfo = [(37.620556, -122.374177), "San Francisco", "SFO"]
sin = [(1.366501, 103.992142), "Singapore", "SIN"]
syd = [(-33.940679, 151.174708), "Sydney", "SYD"]

#create longest flight airport pairs
ewrsin = [ewr[0], sin[0]]
akldoh = [akl[0], doh[0]]
lhrper = [lhr[0], per[0]]
akldxb = [akl[0], dxb[0]]
laxsin = [lax[0], sin[0]]
iahsyd = [iah[0], syd[0]]
dfwsyd = [dfw[0], syd[0]]
jfkmnl = [jfk[0], mnl[0]]
sfosin = [sfo[0], syd[0]]
atljnb = [atl[0], jnb[0]]


#create list of longest flight pairs
pairs = [ewrsin, akldoh, lhrper, akldxb, laxsin, iahsyd, dfwsyd, jfkmnl, sfosin, atljnb]

#creat list of airports
airports = [akl, atl, dfw, doh, dxb, ewr, iah, jfk, jnb, lax, lhr, mnl, per, sfo, sin, syd]

#initialize map
m = folium.Map(location=dxb[0], tiles='Stamen Terrain', zoom_start=2)

#add longest flights to map
for pair in pairs:
    folium.PolyLine(pair).add_to(m)
    
#add airport pins
for airport in airports:
    folium.Marker(
        airport[0], 
        popup=airport[1], 
        tooltip=airport[2], 
        icon=folium.Icon(color='red', icon = 'plane')).add_to(m)
    
#display map
m

#Clearly this presents several issues, since folium just draws 
#straight lines rather than the shortest line between two points 
#(the great circle line). Also, it's 2020; why is anyone still 
#using Mercator projections?
```
