```python
#import pandas
import pandas as pd

#read in data with FIPS etc as strings
df = pd.read_csv('R12653028_SL140.csv',
    dtype=
    {
        'Geo_FIPS':str,
        'Geo_STATE':str,
        'Geo_COUNTY': str
    }
)
```

```python
df.head()
```

```python
#show me all the data
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

#show me a sample
df.sample(5)
```

```python
# look at the data types
# verbose prevents truncating
df.info(verbose=True, null_counts=True)
```

```python
#drop columns with all null values
df = df.dropna(axis=1,how="all")
```

```python
df.info()
```

```python
# list of additional columns to drop
columns_to_drop = ['Geo_GEOID','Geo_STUSAB','Geo_SUMLEV','Geo_GEOCOMP','Geo_FILEID','Geo_LOGRECNO']

# next, drop them!
df = df.drop(columns_to_drop,axis=1)
df.head()
```

```python
#get column list
columns = list(df) # this is the same as df.columns.to_list()
columns
```

```python
df.columns = ['FIPS',
 'Geo_NAME',
 'Geo_QName',
 'Geo_STATE',
 'Geo_COUNTY',
 'Geo_TRACT',
 'Median Gross Rent',
 'Workers',                  # Workers 16 Years and Over:
 'Workers Car',              # Workers 16 Years and Over: Car, Truck, or Van
 'Workers Drove Alone',      # Workers 16 Years and Over: Drove Alone
 'Workers Carpool',          # Workers 16 Years and Over: Carpooled
 'Workers Transit',          # Workers 16 Years and Over: Public Transportation (Includes Taxicab)
 'Workers Motorcycle',       # Workers 16 Years and Over: Motorcycle
 'Workers Bicycle',          # Workers 16 Years and Over: Bicycle
 'Workers Walked',           # Workers 16 Years and Over: Walked
 'Workers Other',            # Workers 16 Years and Over: Other Means
 'Workers Worked at Home',   # Workers 16 Years and Over: Worked At Home
 '% Workers Car',
 '% Workers Drove Alone',
 '% Workers Carpool',
 '% Workers Transit',
 '% Workers Motorcycle',
 '% Workers Bicycle',
 '% Workers Walked',
 '% Workers Other',
 '% Workers Worked at Home']
```

```python
#check new column names
df.head()
```

```python
# get Median Rent stats
df['Median Gross Rent'].describe()
```

```python
# plot a histogram of median rents with 20 bins
df['Median Gross Rent'].plot.hist(bins=20)
```

```python
#import geopandas
import geopandas as gpd

#inport census tracts from LA Times
tracts=gpd.read_file('census-tracts-2012.geojson')

# we only really need FIPS and geometry, so let's subset the data
tracts = tracts[['name','geometry']]

# rename the columns
tracts.columns = ['FIPS','geometry']
```

```python
#merge census tract data with other data
tracts_data=tracts.merge(df,on="FIPS")
```

```python
#make a map of median rents
tracts_data.plot(figsize=(18,18),
                 column='Median Gross Rent',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
#Where can I live for under $1000?
tracts_data[tracts_data['Median Gross Rent'] <=1000].plot(figsize=(18,18),
                 column='Median Gross Rent',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
#Where do people bike to work?
tracts_data.plot(figsize=(18,18),
                 column='% Workers Bicycle',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
#Where do people take transit to work?
tracts_data.plot(figsize=(18,18),
                 column='% Workers Transit',
                 legend=True, 
                 scheme='NaturalBreaks')
```
